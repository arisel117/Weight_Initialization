```python
import numpy as np

from init_weight import draw_histogram

seed = 42
sample_shape = (512, 500)

np.random.seed(seed)
x = np.random.randn(*sample_shape)
```

```python
# Zero init
draw_histogram(x, 0. , 'sigmoid')
draw_histogram(x, 0. , 'relu')
```

```python
# 표준 편차가 0.01인 정규 분포로 init
draw_histogram(x, 0.01 , 'sigmoid')
draw_histogram(x, 0.01 , 'relu')
```

```python
# 표준 편차가 0.3인 정규 분포로 init
draw_histogram(x, 0.3 , 'sigmoid')
draw_histogram(x, 0.3 , 'relu')
```

```python
# Lecun init
draw_histogram(x, 'lecun' , 'sigmoid')
draw_histogram(x, 'lecun' , 'relu')
```

```python
# Xavier init
draw_histogram(x, 'xavier' , 'sigmoid')
draw_histogram(x, 'xavier' , 'relu')
```

```python
# He init
draw_histogram(x, 'he' , 'sigmoid')
draw_histogram(x, 'he' , 'relu')
```
